```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
fig =plt.figure()
ax =plt.axes()
x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x));
```

```python
fig =plt.figure()
ax =plt.axes()
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x))
plt.show()
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```python
plt.plot(x,np.sin(x), color="blue")
```

```python
plt.plot(x,np.sin(x),linestyle='-')
plt.plot(x,np.cos(x),linestyle='--')
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.xlim(0,10)
plt.ylim(-2,2)
```

```python
plt.plot(x,np.sin(x))
plt.axis([-1,11,-1.5,1.5]);
```

```python
plt.plot(x,np.sin(x))
plt.axis('tight');
```

```python
plt.plot(x,np.sin(x))
plt.axis('equal');
```

```python
plt.plot(x,np.sin(x))
plt.title("a sin curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
```
